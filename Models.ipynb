```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.exceptions import ConvergenceWarning
import shap
import warnings

warnings.filterwarnings("ignore", category=ConvergenceWarning)

#Filepath
datasets = {
    'Alpha': "Alpha_with_totalNDE.csv",
    'Beta': "Beta_with_totalNDE.csv",
    'Delta': "Delta_with_totalNDE.csv",
    'Gamma1': "Gamma1_with_totalNDE.csv",
    'Gamma2': "Gamma2_with_totalNDE.csv",
    'Theta': "Theta_with_totalNDE.csv"
}

# Binaryclassification
nde_labels = [0, 1]
target_names = ['Low NDE', 'High NDE']

def bin_total_nde(y):
    return (y > y.median()).astype(int)

# Loop through datasets
for name, file in datasets.items():
    print(f"\n Leave-One-Out Evaluation for {name} dataset")
    df = pd.read_csv(file)

    X = df.iloc[:, :-1]
    y = df.iloc[:, -1]
    y_binned = bin_total_nde(y)

    # Removes NaNs
    X = X[y_binned.notna()]
    y_binned = y_binned[y_binned.notna()]

    if len(y_binned.unique()) < 2:
        print(f"‚ö†Ô∏è {name}: Only one class present ‚Äî skipping.")
        continue

    # Scale input features for MLP andLogistic Regression
    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

    loo = LeaveOneOut()
    results = {
        'Random Forest': {'y_true': [], 'y_pred': [], 'y_prob': []},
        'Logistic Regression': {'y_true': [], 'y_pred': [], 'y_prob': []},
        'Dummy': {'y_true': [], 'y_pred': [], 'y_prob': []},
        'MLP': {'y_true': [], 'y_pred': [], 'y_prob': []}
    }

    for train_idx, test_idx in loo.split(X_scaled):
        X_train, X_test = X_scaled.iloc[train_idx], X_scaled.iloc[test_idx]
        y_train, y_test = y_binned.iloc[train_idx], y_binned.iloc[test_idx]

        # Random Forest
        rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
        rf.fit(X_train, y_train)
        results['Random Forest']['y_true'].append(y_test.values[0])
        results['Random Forest']['y_pred'].append(rf.predict(X_test)[0])
        results['Random Forest']['y_prob'].append(rf.predict_proba(X_test)[0][1])

        # Logistic Regression
        logreg = LogisticRegression(C=1, solver='liblinear', class_weight='balanced', max_iter=1000, random_state=42)
        logreg.fit(X_train, y_train)
        results['Logistic Regression']['y_true'].append(y_test.values[0])
        results['Logistic Regression']['y_pred'].append(logreg.predict(X_test)[0])
        results['Logistic Regression']['y_prob'].append(logreg.predict_proba(X_test)[0][1])

        # Dummy Classifier
        dummy = DummyClassifier(strategy='uniform', random_state=42)
        dummy.fit(X_train, y_train)
        results['Dummy']['y_true'].append(y_test.values[0])
        results['Dummy']['y_pred'].append(dummy.predict(X_test)[0])
        results['Dummy']['y_prob'].append(dummy.predict_proba(X_test)[0][1])

        # MLP Classifier
        mlp = MLPClassifier(hidden_layer_sizes=(25,), alpha=0.001, learning_rate='adaptive',
                            max_iter=2000, random_state=42)
        mlp.fit(X_train, y_train)
        results['MLP']['y_true'].append(y_test.values[0])
        results['MLP']['y_pred'].append(mlp.predict(X_test)[0])
        results['MLP']['y_prob'].append(mlp.predict_proba(X_test)[0][1])

    # Results
    for model_name, res in results.items():
        print(f"\n {model_name} Report for {name}:")
        print(classification_report(res['y_true'], res['y_pred'], labels=nde_labels,
                                    target_names=target_names, zero_division=0))

        cm = confusion_matrix(res['y_true'], res['y_pred'], labels=[0, 1])
        sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
        plt.title(f'{name} | {model_name} (LOOCV) Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.tight_layout()
        plt.show()

    # ROC curves
    plt.figure(figsize=(8, 7))
    for model_name, res in results.items():
        try:
            fpr, tpr, _ = roc_curve(res['y_true'], res['y_prob'])
            auc_score = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc_score:.2f})')
        except ValueError:
            print(f"‚ö†Ô∏è {model_name}: Cannot compute ROC curve.")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{name} | LOOCV ROC Curves')
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show()

    # Logistic Regression (SHAP)
    logreg_final = LogisticRegression(C=1, solver='liblinear', class_weight='balanced', max_iter=1000, random_state=42)
    logreg_final.fit(X_scaled, y_binned)

    explainer = shap.LinearExplainer(logreg_final, X_scaled, feature_perturbation="interventional")
    shap_values = explainer.shap_values(X_scaled)
    shap_df = pd.DataFrame(shap_values, columns=X_scaled.columns)
    shap_df.to_csv(f"{name}_logreg_shap_values.csv", index=False)
    print(f" SHAP values saved to {name}_logreg_shap_values.csv")

    mean_abs_shap = shap_df.abs().mean().sort_values(ascending=False)
    mean_abs_shap.to_csv(f"{name}_logreg_mean_abs_shap.csv", header=["MeanAbsSHAP"])
    print(f" Mean absolute SHAP values saved to {name}_logreg_mean_abs_shap.csv")

    # Random Forest (Feature Importances)
    rf_final = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
    rf_final.fit(X_scaled, y_binned)

    rf_importances = rf_final.feature_importances_
    rf_df = pd.DataFrame({
        'Feature': X_scaled.columns,
        'Importance': rf_importances
    }).sort_values(by='Importance', ascending=False)
    rf_df.to_csv(f"{name}_rf_feature_importance.csv", index=False)
    print(f"Raw RF feature importances saved to {name}_rf_feature_importance.csv")

    # Mean importance per brain region 
    rf_df['Region'] = rf_df['Feature'].apply(lambda x: x.split('.')[0])
    region_mean_importance = rf_df.groupby('Region')['Importance'].mean().sort_values(ascending=False)
    region_mean_importance.to_csv(f"{name}_rf_region_mean_importance.csv", header=["MeanImportance"])
    print(f" Region-mean RF importances saved to {name}_rf_region_mean_importance.csv")
```
